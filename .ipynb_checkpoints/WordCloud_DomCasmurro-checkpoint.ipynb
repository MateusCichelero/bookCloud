Este notebook tem por objetivo descrever os procedimentos para análise visual do livro Dom Casmurro, clássico da literatura brasileira. A técnica escolhida é a wordcloud, onde as informações do livro todo são sintetizadas de forma visual, destacando assim os termos mais frequentes na 'nuvem de palavras'.

A análise se baseia nos seguintes passos:
- Obtenção do texto da obra: foi utilizado como fonte o site de domínio público <a href="https://www.gutenberg.org">Project Gutenberg</a>, onde o texto foi lido diretamente em formato html através da biblioteca requests;
- Processamento do texto: a biblioteca BeautifulSoup foi utilizada para que somente o texto correspondente à narrativa seja considerado na análise.
- Criação da Word Cloud dos termos mais frequentes.

Importando as bibliotecas:

```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from bs4 import BeautifulSoup
```

Salvando a página HTML correspondente ao livro diretamente do site:

```python
url = 'http://www.gutenberg.org/files/55752/55752-h/55752-h.htm'
res = requests.get(url)
html_page = res.content
```

O Beautiful Soup é utilizado para extrair o conteúdo em texto proveniente do HTML: 

```python
soup = BeautifulSoup(html_page, 'html.parser')
```

```python
text = soup.find_all(text=True)
```

Contudo, nem todo o conteúdo extraído é relevante para a análise. Ainda restam algumas tags provenientes do HTML. Com os comandos a seguir é possível verificar quais são elas e definir uma lista de exclusão, para não considerá-las futuramente.

```python
set([t.parent.name for t in text])
```

```python
book = ''
blacklist = [
    '[document]',
    'header',
    'html',
    'meta',
    'head', 
    'input',
    'style',
    'script',
]

for t in text:
    if t.parent.name not in blacklist:
        book += '{} '.format(t)
```

Com o texto pronto para análise, é possível conferir quantos caracteres temos no total:

```python
len(book)
```

Para a criação da nuvem de palavras, basta utilizar as funções da biblioteca wordcloud, que são bem intuitivas. Primeiramente, uma lista de stopwords é criada contendo as palavras que não serão consideradas. Por fim, basta gerar a imagem a partir do texto obtido.

```python
# lista de stopword
stopwords = set(STOPWORDS)
stopwords.update(
    ["da", "meu", "em", "você", "de", "ao", "os","o","a", "não", "mas", "ella","elle","que","para","hoje","também","um","uma",
    "se","era","eu","lhe","como","mais","outro","foi","na","no","porque","dos","das","do","da","por","nem","ou","tambem",
    "quando","só","disse","depois","já"])

# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=800).generate(book)

# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud);
wordcloud.to_file("casmurro_wordcloud.png")
```

Ainda é possível alterar a imagem gerada utilizando uma figura como plano de fundo. O procedimento é descrito abaixo, lembrando que o fundo preto na imagem é que será considerado para alocação das palavras. O resultado é muito interessante, uma ótima sugestão é a criação desse tipo de imagem para falas em filmes, rendendo belos quadros para decoração.

```python
# endereço local da imagem
mask = np.array(Image.open("silhueta.png"))
   
# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1000, height=1000, max_words=2000,
                      mask=mask, max_font_size=200,
                      min_font_size=1).generate(book)

# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud)
wordcloud.to_file("capitu.png")
```
